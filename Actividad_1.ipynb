```python
import pandas as pd
df = pd.read_csv('housing.csv')
df.head(5)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

cols = ['population','households','median_house_value','total_bedrooms','total_rooms','median_income','longitude','latitude']
sns.pairplot(df[cols])
plt.show
```

```python
sns.heatmap(df[cols].corr(), annot=True)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression

X = df['median_income'].values.reshape (-1,1)
Y = df['median_house_value'].values.reshape(-1,1)

ss_X = StandardScaler()
ss_Y = StandardScaler()

X_ss = ss_X.fit_transform(X)
Y_ss = ss_Y.fit_transform(Y)

lr = LinearRegression()
lr.fit (X_ss,Y_ss)
```

```python
import numpy as np

valor_medio_casa = float(input("Ingrese el numero que desee"))
valor_medio_casa_std = ss_X.transform([[valor_medio_casa]])

result_std = lr.predict(valor_medio_casa_std)
result = ss_Y.inverse_transform(result_std)

print('El precio promedio de uan casa segun el valor promedio de una casa es de US$:' , round (result[0][0],2))
```
